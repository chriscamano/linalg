### 1.1.1 $\mathbb{R}^n$
Plotting $\mathbb{R}^n$
In the context of a Jupyter notebook, to visualize $\mathbb{R}^n$ (the $\mathrm{n}$-dimensional real number space), you would typically plot a subset of this space, like $\mathbb{R}^2$ or $\mathbb{R}^3$, since higher dimensions are challenging to represent visually.

- The set $\mathbb{R}^2$, which you can think of as a plane, is the set of all ordered pairs of real numbers:
$$
\mathbb{R}^2=\{(x, y): x, y \in \mathbb{R}\}
$$

```python
class R2Space(Scene):
    def construct(self):
        plane = NumberPlane(x_range=[-10, 10], y_range=[-10, 10])
        plane.add_coordinates()
        title = MathTex("\mathbb{R}^2")
        title.to_corner(UR)  # Position title to the upper right corner
        self.add(plane, title)
%manim -ql  -v WARNING --disable_caching R2Space
```

### From ordered pairs  to vectors  
The  concept of ordered pairs can be extended to vectors. A vector in $\mathbb{R}^2$ essentially represents a directed line segment from the origin to a point in the plane. In this context, you can represent a vector as:
$$
\mathbf{v}=\left[\begin{array}{l}
x \\
y
\end{array}\right]
$$

Here, $\mathbf{v}$ symbolizes the vector, and it encapsulates both the magnitude and direction.


```python
class R2Space(Scene):
    def construct(self):
        # Create a 2D grid
        plane = NumberPlane(x_range=[-10, 10,1], y_range=[-10, 10,1])

        # Create a point at (2, 3)
        point = Dot(point=plane.c2p(2, 3), color=BLUE)
        point_label = MathTex("(2, 3)").next_to(point, UR)

        # Create a vector from the origin to (2, 3)
        vector = Arrow(start=ORIGIN, end=plane.c2p(2, 3), color=GREEN)
        vector_label = MathTex("\\mathbf{v} = \\begin{bmatrix} 2 \\\\ 3 \\end{bmatrix}").next_to(vector, UR)

        # Animate the point becoming a vector
        self.play(Create(plane), Create(point),Write(point_label))
        self.wait(1)
        self.play(FadeOut(point_label), Create(vector), Write(vector_label))
        self.wait(2)
%manim -qh  -v WARNING --disable_caching R2Space
```

### 1.1.1 $\mathbb{C}^n$
The set of all complex numbers is denoted by $\mathbb{C}$ :
$$
\mathbb{C}=\{a+b i: a, b \in \mathbb{R}\} .
$$
If $a \in \mathbb{R}$, we identify $a+0 i$ with the real number $a$. Thus we think of $\mathbb{R}$ as a subset of $\mathbb{C}$. We usually write $0+b i$ as just $b i$, and we usually write $0+1 i$ as just $i$.

```python
class ComplexPlaneExample(Scene):
    def construct(self):
        complex_plane = ComplexPlane(x_range=[-10, 10], y_range=[-10, 10])
        complex_plane.add_coordinates()
        title = MathTex("\mathbb{C}")  # The symbol for the complex plane
        title.to_corner(UR)  # Position title to the upper right corner
        self.add(complex_plane, title)
%manim -ql  -v WARNING --disable_caching ComplexPlaneExample
```

### From complex numbers to vectors  
A nice property about the complex plane is that values of the form $a+bi$ have a very natural represetation, for example the complex number $2+3i$ can be represented as a vector as follow 


```python
class ComplexPlaneVector(Scene):
    def construct(self):
        # Create a complex plane
        complex_plane = ComplexPlane(x_range=[-10, 10, 1], y_range=[-10, 10, 1])
        complex_plane.add_coordinates()
        title = MathTex("\mathbb{C}")
        title.to_corner(UR) 
        # Choose a point in the complex plane, for example, 2 + 3i
        point = complex_plane.n2p(2 + 3j)  # Converts the complex number to a point in the plane

        # Create a vector to this point
        vector = Arrow(start=ORIGIN, end=point, buff=0, color=RED)

        # Label for the point
        point_label = MathTex("2 + 3i").next_to(vector.get_end(), UR)

        # Animation
        self.add(complex_plane,title)
        self.play(Create(vector), Write(point_label))
        self.wait(1)
%manim -qh  -v WARNING --disable_caching ComplexPlaneVector
```
